```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,8))
grafica = sns.histplot(data=inmueble, x="Precio_millon", kde=True, hue="Tipo" )
grafica.set_title("Distribucion de valores de inmuebles en Bogota")
plt.xlim((50,1000))
plt.savefig("/content/drive/MyDrive/Colab Notebooks/inemrcion de datos/grafico0.1",format="png")
plt.show()
```

```python
inmueble["Valor_m2_millones"] = inmueble["Precio_millon"]/inmueble["Area"]
inmueble.head()
```

```python
inmueble.groupby("Barrio").mean()
```

```python
datos_barrio = inmueble.groupby("Barrio").sum()
datos_barrio
```

```python
datos_barrio["Valor_m2_barrio"] = datos_barrio["Precio_millon"]/datos_barrio["Area"]
datos_barrio
```

```python
m2_barrio = dict(datos_barrio["Valor_m2_barrio"])
```

```python
inmueble["Valor_m2_barrio"] = inmueble["Barrio"].map(m2_barrio)
inmueble.head(5)
```

```python
top_barrio = inmueble['Barrio'].value_counts()[:10].index
```

```python
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
datos_barrio.query("Barrio in @top_barrio")
```

```python
plt.figure(figsize=(10,8)) 
ax = sns.boxplot(x="Barrio", y="Valor_m2_millones", data = datos_barrio.query("Barrio in @top_barrio"))
ax.tick_params(axis="x", rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6)) 
ax = sns.boxplot(x="Barrio", y="Valor_m2_millones", data = datos_barrio.query("Barrio in @top_barrio y Valor_m2_millones < 15 "))
ax.tick_params(axis="x", rotation=45)
plt.show()
```
